# Keras.models:

# 1. Sequential: 一層一層往上堆疊

# 2. Model: 任意層數往上堆疊 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense # 全連接層

layers = [
  # 784*128(線條) + 128(bias) = 100480
  Dense(128, activation='relu', input_dim=784), # 28*28
  # 128*10(線條) + 10(bias) = 1290
  Dense(10, activation='softmax') # softmax --> m=1(y=x)
]
model = Sequential(layers)
model.summary()
```
